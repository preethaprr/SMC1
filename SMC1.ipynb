```python
import os
import numpy as np
import struct
import imageio


def run_length_encode(image):
    """
    Run Length Encoding (RLE) for image compression.
    """
    encoded_data = []
    current_value = None
    count = 0

    for pixel in image.flatten():
        if pixel != current_value:
            if current_value is not None:
                encoded_data.append((current_value, count))
            current_value = pixel
            count = 1
        else:
            count += 1

    encoded_data.append((current_value, count))
    return encoded_data


def lzw_compress(data):
    """
    Lempel-Ziv-Welch (LZW) compression algorithm.
    """
    dictionary_size = 256
    dictionary = {bytes([i]): i for i in range(dictionary_size)}
    compressed_data = bytearray()
    current_sequence = b""

    for byte in data:
        sequence_plus_byte = current_sequence + bytes([byte])
        if sequence_plus_byte in dictionary:
            current_sequence = sequence_plus_byte
        else:
            compressed_data += struct.pack('<I', dictionary[current_sequence])
            dictionary[sequence_plus_byte] = dictionary_size
            dictionary_size += 1
            current_sequence = bytes([byte])

    if current_sequence in dictionary:
        compressed_data += struct.pack('<I', dictionary[current_sequence])

    return compressed_data


def main():
    # Folder path containing images
    folder_path = "/content/train"
    output_folder = "/content/compressed"

    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    total_original_size = 0
    total_compressed_size = 0

    # Iterate over each file in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            # Load the image
            image_path = os.path.join(folder_path, filename)
            image = imageio.imread(image_path)

            print(f"Processing image: {filename}")

            # Run Length Encoding
            encoded_data = run_length_encode(image)

            # LZW Compression
            values = np.array([pixel for pixel, _ in encoded_data], dtype=np.uint8)
            compressed_values = lzw_compress(values)

            # Check if compression is successful
            if len(compressed_values) < image.nbytes:
                print("Successful image compression")
            else:
                print("Failed to compress image")

            # Save compressed data to file
            output_file = os.path.join(output_folder, filename.replace(".", "_compressed."))
            with open(output_file, "wb") as f:
                f.write(compressed_values)

            # Update total sizes
            total_original_size += image.nbytes
            total_compressed_size += len(compressed_values)

    # Print overall compression size and ratio
    print("Overall compression:")
    print(f"Total original size: {total_original_size} bytes")
    print(f"Total compressed size: {total_compressed_size} bytes")
    overall_ratio = total_original_size / total_compressed_size
    print(f"Overall compression ratio: {overall_ratio:.2f}")


if __name__ == "__main__":
    main()
```

```python
import os
import math

def calculate_entropy(file_path):
    """
    Calculate the entropy of a file.
    """
    # Read the contents of the file
    with open(file_path, 'rb') as f:
        data = f.read()

    # Calculate the frequency of each byte
    frequency = {}
    total_bytes = len(data)
    for byte in data:
        if byte in frequency:
            frequency[byte] += 1
        else:
            frequency[byte] = 1

    # Calculate entropy
    entropy = 0
    for count in frequency.values():
        probability = count / total_bytes
        entropy -= probability * math.log2(probability)

    return entropy

# Iterate over each file in the output folder
for filename in os.listdir(output_folder):
    file_path = os.path.join(output_folder, filename)
    if os.path.isfile(file_path):
        entropy = calculate_entropy(file_path)
        print(f"Entropy of {filename}: {entropy:.2f}")
```

```python
import hashlib
import json
import time

class Block:
    def __init__(self, index, timestamp, data, previous_hash):
        self.index = index
        self.timestamp = timestamp
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        return hashlib.sha256(str(self.index).encode() +
                              str(self.timestamp).encode() +
                              str(self.data).encode() +
                              str(self.previous_hash).encode()).hexdigest()
```

```python
class Blockchain:
    def __init__(self):
        self.chain = [self.create_genesis_block()]

    def create_genesis_block(self):
        # Manually create the first block (genesis block)
        return Block(0, time.time(), "Genesis Block", "0")

    def add_block(self, new_block):
        new_block.previous_hash = self.chain[-1].hash
        new_block.hash = new_block.calculate_hash()
        self.chain.append(new_block)

def store_encrypted_data_in_blockchain(encrypted_data, ipfs_hash, blockchain):
    # Create a new block containing encrypted data and its IPFS hash
    new_block = Block(len(blockchain.chain), time.time(), {"encrypted_data": encrypted_data, "ipfs_hash": ipfs_hash}, blockchain.chain[-1].hash)
    blockchain.add_block(new_block)

def main():
    blockchain = Blockchain()

    # Encrypted data obtained after encryption
    encrypted_data = b'\x12\x34\x56\x78\x9a\xbc\xde\xf0'  # Example encrypted data

    # Generate IPFS hash for encrypted data
    ipfs_hash = generate_ipfs_hash(encrypted_data)

    # Store encrypted data and IPFS hash in the blockchain
    store_encrypted_data_in_blockchain(encrypted_data, ipfs_hash, blockchain)

    print("Encrypted data and IPFS hash stored securely in the blockchain.")

if __name__ == "__main__":
    main()
```

```python
import os
import time
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.kdf.pbkdf2 import PBKDF2HMAC
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives import serialization
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes


def encrypt_data(data, key, nonce, output_folder):
    """
    Encrypt data using symmetric encryption (AES-GCM).
    """
    cipher = Cipher(algorithms.AES(key), modes.GCM(nonce), backend=default_backend())
    encryptor = cipher.encryptor()
    ciphertext = encryptor.update(data) + encryptor.finalize()

    # Save encrypted data to file
    output_file = os.path.join(output_folder, "encrypted_" + os.path.basename(output_folder) + ".bin")
    with open(output_file, "wb") as f:
        f.write(nonce)
        f.write(encryptor.tag)
        f.write(ciphertext)

    return output_file


def decrypt_data(encrypted_data, key, nonce, tag):
    """
    Decrypt data using symmetric decryption (AES-GCM).
    """
    try:
        cipher = Cipher(algorithms.AES(key), modes.GCM(nonce, tag), backend=default_backend())
        decryptor = cipher.decryptor()
        start_time = time.time()
        decrypted_data = decryptor.update(encrypted_data) + decryptor.finalize()
        decryption_time = time.time() - start_time
        return decrypted_data, decryption_time, True
    except Exception as e:
        return str(e), 0, False


def main(output_folder):
    # Derive encryption key from passphrase
    passphrase = b'secret_passphrase'
    salt = os.urandom(16)
    kdf = PBKDF2HMAC(
        algorithm=hashes.SHA256(),
        length=32,
        salt=salt,
        iterations=100000,
        backend=default_backend()
    )
    key = kdf.derive(passphrase)

    # Generate a random nonce
    nonce = os.urandom(16)

    total_decryption_time = 0

    # Iterate over each file in the output folder
    for filename in os.listdir(output_folder):
        file_path = os.path.join(output_folder, filename)
        if os.path.isfile(file_path):
            with open(file_path, "rb") as f:
                data = f.read()

            print(f"Encrypting file: {filename}")

            # Encrypt file contents using symmetric encryption
            encrypted_file_path = encrypt_data(data, key, nonce, output_folder)

            print(f"Encryption complete: {filename} -> {encrypted_file_path}")

            # Now decrypt the encrypted file for verification
            with open(encrypted_file_path, "rb") as f:
                nonce = f.read(16)
                tag = f.read(16)
                encrypted_data = f.read()

            decrypted_data, decryption_time, success = decrypt_data(encrypted_data, key, nonce, tag)
            if success:
                print(f"Decryption successful for file {filename}")
                print(f"Decryption time: {decryption_time} seconds")
                total_decryption_time += decryption_time
            else:
                print(f"Decryption failed for file {filename}: {decrypted_data}")

    print(f"Total decryption time: {total_decryption_time} seconds")


# Define the output folder path
output_folder = "/content/compressed"

# Call the main function with the output_folder variable
main(output_folder)
```

```python
import os
from shutil import copyfile
import time

def decompress_images(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)

    total_decryption_time = 0
    num_images = 0

    for filename in os.listdir(input_folder):
        if filename.endswith("_compressed.jpg") or filename.endswith("_compressed.png"):
            compressed_file_path = os.path.join(input_folder, filename)
            output_file_path = os.path.join(output_folder, filename.replace("_compressed", ""))

            start_time = time.time()

            # Decompression process
            copyfile(compressed_file_path, output_file_path)

            end_time = time.time()
            decryption_time = end_time - start_time
            total_decryption_time += decryption_time
            num_images += 1

            print(f"Decompressed: {filename}, Decryption Time: {decryption_time:.4f} seconds")

    if num_images > 0:
        avg_decryption_time = total_decryption_time / num_images
        print(f"Average Decompressed Time: {avg_decryption_time:.4f} seconds")

if __name__ == "__main__":
    input_folder = "/content/compressed"
    output_folder = "/content/decompressed"
    decompress_images(input_folder, output_folder)
```
